```python
import torch

s = torch.tensor(0,dtype=torch.float32)
for i in range(1000):
    s += torch.tensor(0.01,dtype=torch.float32)
print(s)

s = torch.tensor(0,dtype=torch.float16)
for i in range(1000):
    s += torch.tensor(0.01,dtype=torch.float16)
print(s)

s = torch.tensor(0,dtype=torch.float32)
for i in range(1000):
    s += torch.tensor(0.01,dtype=torch.float16)
print(s)

s = torch.tensor(0,dtype=torch.float32)
for i in range(1000):
    x = torch.tensor(0.01,dtype=torch.float16)
    s += x.type(torch.float32)
print(s)
```

```python
import torch.nn as nn 
class ToyModel(nn.Module):
    def __init__(self, in_features: int, out_features: int):
        super().__init__()
        self.fc1 = nn.Linear(in_features, 10, bias=False)
        self.ln = nn.LayerNorm(10)
        self.fc2 = nn.Linear(10, out_features, bias=False)
        self.relu = nn.ReLU()
    
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.ln(x)
        x = self.fc2(x)
        return x

model = ToyModel(10, 10)
print(f"parameters types!")
print(f"model.fc1.weight.dtype: {model.fc1.weight.dtype}")
print(f"model.fc2.weight.dtype: {model.fc2.weight.dtype}")
print(f"model.ln.weight.dtype: {model.ln.weight.dtype}")
print(f"model.ln.bias.dtype: {model.ln.bias.dtype}")
```

上面模型有参数只有fc1, ln和fc2
fc1和fc2都是矩阵乘法，所以使用fp16，因为矩阵里面的值都是0点几，所以乘以之后变成较小的数字，不像梯度那么小，所以使用fp16
但是对于layernorm，需要求和来获取均值，所以使用fp32
